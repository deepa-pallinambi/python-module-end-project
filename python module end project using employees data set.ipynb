```python
print("\n Age group of most of the employees: \n")
df["Age"].value_counts()
bins=[20,25,30,35,40,45,50]
#using cut function to categorise employees into different age groups
age_groups=pd.cut(df["Age"],bins=bins)
print(age_groups.value_counts())
age_count=age_groups.value_counts()
max_age=age_count.idxmax()
print("age group of most of the employees is",max_age)
```

```python
df[["Team","Position","Salary"]].sort_values(by="Salary",ascending=False).head(3)
```

```python
highest_team_position=df.groupby(["Team","Position"])["Salary"].sum().idxmax()
print("highest team and position with respect to spending of salary",highest_team_position)
```

```python
correlation=df["Age"].corr(df["Salary"])
print("correlation between Age and Salary is",correlation)
print("this is weak positive correlation ranges between 0.2-0.39")
```

```python
print("\n Visual Representation: \n")
sns.scatterplot(data=df,x="Age",y="Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("correlation between age and salary")
plt.show()
```
